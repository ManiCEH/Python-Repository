## Problem# 1: Write a program to open the lock.

```python
lock = 999
#initialization
combination = 000
#condition
while combination != lock:
  # print(combination)
  combination += 1 #updation
print("Lock is opened with combination,",combination)  
```

## Problem #5: Level 2

Multiple input from user and print the multiplication table.

```python
T = int(input('how many table you want to ask:'))
i = 1 
while i<=T:
  n = int(input("tell me the table of :"))
  j = n
  while j<=n*10:
    print(j)
  
    j = j+n
  i = i+1  
```
